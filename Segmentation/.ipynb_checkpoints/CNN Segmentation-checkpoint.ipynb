A unsupervised segmentation approach is tested, based on the paper "UNSUPERVISED IMAGE SEGMENTATION BY BACKPROPAGATION"  by Asako Kanezaki. The key differences are that in this notebook, a the network is pretrained with the two first layers of VGG19, and that a fixed number of 2 clusters is defined: background and items.

```python
K = 10000 #number of superpixels (max)
max_clusters = 10
```

```python
#input image:
img = layers.Input((2464//sf,3280//sf,3),name = 'img')
#input superpixel: 

#take the feature extractor:

features = feature_ex_ref(img)
#features = feature_ex_inc(img)         

#get the initial size from the image

### for VGG19:
fullscale = layers.UpSampling2D(name='revmaxpool')(features)
#combine the channels:
cluster = layers.Conv2D(max_clusters,1,name = 'combiner')(fullscale)

###for inceptionResNetv2:
###for inceptionResNetv2:
#strides = [8,8]#[img.shape[1]//(features.shape[1]-1),(img.shape[2])//(features.shape[2]-1)]
#pad = [imgd-stride*featuresd for imgd,stride,featuresd in zip(img.shape[1:3],strides,features.shape[1:3])]
#if pad[0]%2 or pad[1]%2:
#    print("ERROR: Please use another input shape")
#kernel_size = (1,1)
#cluster = layers.Conv2DTranspose(max_clusters,kernel_size,strides=strides, name='combiner')(features)
#padded_cluster = ctools.ReflectionPadding2D(padding =[pad[0]//2,pad[1]//2])(cluster)

clustern = layers.LayerNormalization(name = 'cluster_normalization',axis = (1,2),
                                    scale = False,center = False)(cluster)

#get the clustered image:
#constrain the superpixels:

unsupervised_seg = Model(inputs = img,outputs=clustern)
```

```python
#define a model whose training epoch is divided into discriminator and generator training
class AutoSegmenter(keras.Model):
    def __init__(self, segmenter, inputs):
        super(AutoSegmenter, self).__init__()
        self.segmenter = segmenter
        self.inputs = inputs

    def compile(self, optimizer):
        super(AutoSegmenter, self).compile()
        self.optimizer = optimizer
        self.loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
        
    def train_epoch(self, dataset,limit_sample =None, n_iter = 4,loss_list = []):
        
        for k,data in enumerate(dataset):
            if limit_sample is not None:
                if k == limit_sample:
                    break
            
            img = tf.expand_dims(data[0], 0, name=None)
            superpixels = data[1]
            for iter in range(n_iter):
                
                with tf.GradientTape() as tape:
                    labels = self.segmenter(img)
                    max_label =  tf.math.argmax(labels,axis=-1)
                    corrected_labels = tf.zeros_like(max_label)
                    for i in range(tf.reduce_max(superpixels)):
                        mask = tf.where(superpixels==i,max_label,-1)
                        labels_id, idx, count = tf.unique_with_counts(tf.reshape(mask,[-1]))
                        count_val = tf.where(labels_id == -1,0,count)
                        superpix_maxlabel = labels_id[tf.math.argmax(count_val)]
                        corrected_labels = tf.where(superpixels==i,superpix_maxlabel,corrected_labels)
                    labels_cor, idx = tf.unique(tf.reshape(corrected_labels,[-1]))
                    loss = self.loss(corrected_labels,labels)
                    loss_list.append(loss.numpy())
                    tf.print("loss after "+str(k+1)+" samples: "+str(loss.numpy())+" with "+
                             str(len(labels_cor))+" different clusters   ",end='\r')
                    grads  = tape.gradient(loss, self.segmenter.trainable_weights)
                    #if (i%batchsize == 0 and i !=0):
                        #update the weights
                    self.optimizer.apply_gradients(zip(grads, self.segmenter.trainable_weights))
                    
            #plot the result:
            fig,ax = plt.subplots(1,3,figsize=(18,6))
            img_f = data[0]
            ax[0].imshow(destand(tf.cast(img_f,float),perc = 0.0001))
            ax[1].imshow(tf.cast(tf.math.argmax(labels,axis=3)[0],float),cmap='tab10')
            ax[2].plot(loss_list)
            plt.show() 
        return loss_list
```

```python
auto = AutoSegmenter(inputs=img,segmenter=unsupervised_seg)
auto.compile(optimizer = tf.keras.optimizers.Adam())
```

```python
def find_img_rec(base):
    ret = []
    try:
        for folder in os.listdir(base):
            if folder[-4:]==".jpg":
                ret.append(os.path.join(base,folder))
            else:
                ret+= find_img_rec(os.path.join(base,folder))
    except:
        ret = []
    return ret
base_img = os.path.join('Q:','Eco','dominic')
img_names = find_img_rec(base_img)
base_mean = os.path.join('Q:','Eco','Gabriel','bundles')
```

```python
def standardize(img,mean=None,var=None,epsilon = 0.0001):
    if mean is None:
        mean = np.mean(img)
    meanless = img-mean
    if var is None:
        var = np.var(meanless)
    
    img_n = meanless/(var+epsilon)
    return img_n
def destand(img_n, perc = 1,epsilon = 0.0001):
    pl,ph = np.percentile(img_n, [perc,100-perc])
    img = np.array((img_n-pl)/(epsilon+ph-pl))
    img[img>1]=1
    img[img<0]=0
    return img
```

```python
#data set creator:
def dataset_gen(img_names=img_names,base_img = base_img, base_mean=base_mean,threshold=50):
    n = 0
    np.random.shuffle(img_names)
    while n < len(img_names):
        img = skimage.io.imread(img_names[n])
        #get the mean from the dataset
        headless_path = img_names[n][len(base_img):]
        newhead_path = base_mean+headless_path
        mean_path = os.path.split(newhead_path)[0]
        mean_name = os.path.join(mean_path,'mean_from_aquisition.png')
        mean_img = skimage.io.imread(mean_name)
        img_meanless  = img.astype(int)-mean_img
        if np.percentile(np.abs(img_meanless),99)>threshold:
            #standardize the image
            imgn = standardize(img_meanless[0::sf,0::sf])

            superpixels = slic(imgn,n_segments=K//sf,compactness = 10, sigma  =10,
                                      min_size_factor=0.001, max_size_factor=1,start_label = 0)
            yield(imgn,superpixels)
        n+=1
```

```python
for i,data in enumerate(dataset_gen()):
    if i>1:
        break
    fig,ax = plt.subplots(1,2)
    ax[0].imshow(destand(data[0],perc = 0.0001))
    ax[1].imshow(data[1])
    plt.show()
```
