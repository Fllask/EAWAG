# 1. K-Fold cross validation
In this notebook, the design choices concerning the k-fold cross-validation will be explained.

```python
#importing the useful libraries
#from IPython.display import Markdown, display
import importlib
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import copy
#from scipy import signal
import warnings
#from scipy import sparse
import matplotlib.colors as mcolors
stf_dir = os.path.join(os.pardir, 'spacetimeformer')
sys.path.append(stf_dir)
```

```python
#load the dataset
data_path = "..\\..\\..\\Datasets\\Forecasting_aqua\\data_h_rolling_cst_int_v4.csv"
format = "%Y-%m-%d %H:%M:%S"
df = pd.read_csv(data_path)
time_df = pd.to_datetime(df["Datetime"], format=format)
try:
    df.drop("sample",axis=1,inplace=True)
except:
    pass
df['Datetime']= time_df
#According to the work of Stefanie Merkly, these keys were more important than the others to predict chorophill a
keys_imp = ['ciliate','mean_chla','cv_chla_day','cv_chla_depth',
            'nauplius','Ammonium','Nitrat','mean_schmidt','windspeed_max',
            'mean_thermocline_depth','mean_epi_temp','mean_oxycline_depth',
            'mean_mixed_layer_depth','mean_global_radiation'
           ]
#add some usefull variables at the end of the table
df['year'] = df['Datetime'].dt.year
df['month'] = df['Datetime'].dt.month
df['day'] = df['Datetime'].dt.day
df['hour'] = df['Datetime'].dt.hour
#drop the first datapoints the first months are only meteorological data
first_point = (~df['diatom'].isnull()).idxmax()
dfs = df.drop(index=range(first_point))
df = dfs.set_index('Datetime')
```

## What is k-fold:
K-fold is a method of validation, mainly used for hyper-parameters tuning, where the data is separated into k different parts (or folds). The model is then trained k times on all but one of the parts, while the remaining one is use to validate it and to unsure that no overfitting happen. Note that due to the long time required to train the model k times, the model is then tested with a hold-out approach instead of another k-fold: The test data is never used in this training procedures

```python
#number of folds
k = 3
#delay (in days)
lag = 1
#portion of the data used for the test
test_prop = 0.15

test_len = int(len(df)*test_prop)
dftv = df[:-test_len].copy()

#not used for now 
dftest = df[-test_len:].copy()


#explanatory variables (only the chla is here for this example)
inputs_tv = dftv['mean_chla_depth'][:-24*lag]
# target variables to predict (only the chla is here for this example)
targets_tv = dftv['mean_chla_depth'][24*lag:].reset_index(drop=True)


breakpoints = np.linspace(0,len(inputs_tv),k+1,dtype=int)
fig,axs= plt.subplots(k,1,figsize=(10,k*3),tight_layout=True)
for idx,ax in enumerate(axs):
    ax.set_title(f"fold number {idx}")
    train =ax.scatter(dftv.index[breakpoints[idx+1]:breakpoints[-1]],
                      inputs_tv[breakpoints[idx+1]:breakpoints[-1]],s=1,c='tab:blue')
    train =ax.scatter(dftv.index[breakpoints[0]:breakpoints[idx]],
                      inputs_tv[breakpoints[0]:breakpoints[idx]],s=1,c='tab:blue')
    val = ax.scatter(dftv.index[breakpoints[idx]:breakpoints[idx+1]],
                     inputs_tv[breakpoints[idx]:breakpoints[idx+1]],s=1,c='tab:orange')
    ax.set_ylabel("input variable")
    ax.set_xlabel("Date")
    ax.legend([train,val],['training data','validation data'])
```

To allow an easy comparison between the different models and hyper-parameters, a function was designed to automatically train and verify an arbitrary model, with the only restriction that the model must have a fit and a predict method:

```python
#example of a simple model that consider that each output is the mean value of its corresponding training data
class naive_model:
    #initialize the model
    def __init__(self,target_keys:list):
        from sklearn.preprocessing import StandardScaler
        self.target_keys = target_keys
        self._scalerout = StandardScaler()
    #the fit method argument must be the following 
    def fit(self,inputs,target,breakpoints):
        self._scalerout = self._scalerout.fit(target.values)
    def pred(self,inputs):
        npred = np.zeros((len(inputs),len(self.target_keys)))
        scaled_back_pred = self._scalerout.inverse_transform(npred)
        return scaled_back_pred   
```

The k-fold function directly compute a score for each folds. For now, the metrics of interest are the R2 score, (by target or global), and the mean absolute error (on data normalized over the training set statistics)


```python
def score(y_true,y_pred,mean_y_train, epsilon=1e-16):
    
    MAEn = 1-np.nanmean(np.abs(y_true-y_pred))/np.nanmean(np.abs(y_true-mean_y_train)+epsilon)
    R2_cat = 1-(np.nanmean(np.square(y_true-y_pred),axis=0)/
                np.nanmean(np.square(y_true-mean_y_train),axis=0)+epsilon)
    #note: the aggregation of R2 is quite tricky, as the two variables have highly different variances
    # the chosen method was to simply take the mean of the R2 coefficients of each variables
    R2 = np.nanmean(R2_cat)
    return {'R2':R2,'MAEn':MAEn,'R2_cat':R2_cat}
```

```python
def kfold(model,input_tv,target_tv,n_fold):
    breakpoints = np.linspace(0,len(input_tv),n_fold+1,dtype=int)
    res = []
    for fold in range(n_fold):
        print(f"                                                 Fitting fold n°{fold}        ",end='\r')
        #define the validation split
        input_v = input_tv[breakpoints[fold]:breakpoints[fold+1]]
        y_v  = target_tv[breakpoints[fold]:breakpoints[fold+1]]
        #define the training split
        input_t = pd.concat([input_tv[:breakpoints[fold]],input_tv[breakpoints[fold+1]:]])
        y_tr = pd.concat([target_tv[:breakpoints[fold]],target_tv[breakpoints[fold+1]:]])
        #train the model
        model.fit(input_t,y_tr,[breakpoints[fold],breakpoints[fold+1]])
        #predict both the training data, and the validation
        y_predtr = model.pred(input_t)
        y_predv = model.pred(input_v)
        #store the mean of the training data
        mean_ytrain = np.nanmean(y_tr,axis=0)
        with warnings.catch_warnings():
            warnings.filterwarnings("ignore", category=RuntimeWarning)
            s_train = score(y_tr.to_numpy(),y_predtr,mean_ytrain)
            s_val = score(y_v.to_numpy(),y_predv,mean_ytrain)
        res.append({'model':copy.deepcopy(model),'fold':fold,'bp': breakpoints,'y_predtr':y_predtr, 'y_predv':y_predv,
                    's_train':s_train,'s_val':s_val,'outkeys':target_tv.keys(),'mean_ytrain':mean_ytrain})
    print(" "*100,end='\r')
    return res
```

As an example, here is how the model k-fold validation would look for a 4-fold validation using the naive model to predict the  chlorophyll a concentration one day later:

```python
#chose a lag
lag = 1

#explanatory variables
inputs_tv = dftv[:-24*lag]
# target variables to predict 
target_keys = ['mean_chla_depth']
targets_tv = dftv[24*lag:][target_keys]

#chose a model
model = naive_model(target_keys)

#run the kfold
n_fold = 4
res = kfold(model,inputs_tv,targets_tv,n_fold)
print("Model trained"+' '*70)
```

The results can then be visualized:

```python
#lets only focus on a specific fold
fold_id  = 3
res_fold = res[fold_id]

#and lets only focus on a specific key
key_id  = 0

#plot the results##############
fig,ax = plt.subplots(1,figsize=(15,4))
true_data = targets_tv.iloc[:,key_id]
#plot the true data
true = ax.plot(true_data,c='k')

#plot the predicted training data
bp = res_fold['bp']
train = ax.plot(true_data.index[0:bp[fold_id]],res_fold['y_predtr'][0:bp[fold_id],key_id],c= 'tab:blue')
ax.plot(true_data.index[bp[fold_id+1]:],res_fold['y_predtr'][bp[fold_id]:,key_id],c='tab:blue')
#plot the validatation data
val = ax.plot(true_data.index[bp[fold_id]:bp[fold_id+1]],res_fold['y_predv'][:,key_id],c='tab:orange')

ax.legend([true[0],train[0],val[0]],
         ['True data',
          f'Training predictions (R2={res_fold["s_train"]["R2_cat"][key_id]:.2f})',
          f'Validation predictions (R2={res_fold["s_val"]["R2_cat"][key_id]:.2f})'
         ])
ax.set_ylabel("Concentration [ROI/s]")
ax.set_title(f"fold number {fold_id} results for {target_keys[key_id]}")
plt.show()
```

Note that by definition, this model will always return a R2 score of 0. 

## Why k-fold
This validation method is useful, as it allow a detection of over-fitting, while keeping as much data as possible both in training and validation: The training is done on $\frac{k-1}{k}$ of all the data at each fold, and the final validation covers the whole dataset. 

To illustrate why this is useful, let's take a simple linear model:

```python
class linear_model:
    def __init__(self,target_keys:list):
        from sklearn.linear_model import LinearRegression
        from sklearn.preprocessing import StandardScaler
        self.skmodel = LinearRegression()
        self.target_keys = target_keys
        self._scalerin = StandardScaler()
        self._scalerout = StandardScaler()
    def fit(self,inputs,target,breakpoints):
        inputs = inputs.drop(columns=['Datetime'],errors='ignore')
        self._scalerin = self._scalerin.fit(inputs.values)
        self._scalerout = self._scalerout.fit(target.values)
        ninp = self._scalerin.transform(inputs.values)
        #fill nans with 0s
        ninp = np.nan_to_num(ninp)
        ntar = self._scalerout.transform(target.values)
        ntar = np.nan_to_num(ntar)
        self.skmodel.fit(ninp,ntar)
    def pred(self,inputs):
        inputs = inputs.drop(columns=['Datetime'],errors='ignore')
        if len(inputs)==0:
            return np.zeros((len(inputs),len(self.target_keys)))
        ninp = self._scalerin.transform(inputs.values)
        ninp = np.nan_to_num(ninp)
        npred = self.skmodel.predict(ninp)
        return self._scalerout.inverse_transform(npred)    
```

and fit it the same way the constant model was (with an increased lag):

```python
#chose a lag
lag = 10

#explanatory variables
inputs_tv = dftv[:-24*lag]
# target variables to predict 
target_keys = ['mean_chla_depth']
targets_tv = dftv[24*lag:][target_keys]

#chose a model
model = linear_model(target_keys)


#run the kfold
n_fold = 4
#run the kfold (while keeping the same parameters as before)
res = kfold(model,inputs_tv,targets_tv,n_fold)
print("Model trained"+' '*70)
```

```python
#lets only focus on a specific fold
fold_id  = 3
res_fold = res[fold_id]

#and lets only focus on a specific key
key_id  = 0

#plot the results##############
fig,ax = plt.subplots(1,figsize=(15,4))
true_data = targets_tv.iloc[:,key_id]
#plot the true data
true = ax.plot(true_data,c='k')

#plot the predicted training data
bp = res_fold['bp']
train = ax.plot(true_data.index[0:bp[fold_id]],res_fold['y_predtr'][0:bp[fold_id],key_id],c= 'tab:blue')
ax.plot(true_data.index[bp[fold_id+1]:],res_fold['y_predtr'][bp[fold_id]:,key_id],c='tab:blue')
#plot the validatation data
val = ax.plot(true_data.index[bp[fold_id]:bp[fold_id+1]],res_fold['y_predv'][:,key_id],c='tab:orange')

ax.legend([true[0],train[0],val[0]],
         ['True data',
          f'Training predictions (R2={res_fold["s_train"]["R2_cat"][key_id]:.2f})',
          f'Validation predictions (R2={res_fold["s_val"]["R2_cat"][key_id]:.2f})'
         ])
ax.set_ylabel("Concentration [ROI/s]")
ax.set_title(f"fold number {fold_id} results for {target_keys[key_id]}")
plt.show()
```

As one can see, the training fit is way better than the validation one. This is the problem known as over-fitting. 

### Why no shuffling
In formal k-fold, the data is shuffled before the splits: this allows to have a more homogeneous validation and training set. The problem with this method is that our models are supposed to be applied to data that they have never seen, and thus the validation should try to emulate this. Due to the quite long correlation times of these real variables, using a shuffled dataset would return a result strongly biased. This is also the reason why the test set was kept as hold-last-out: this is the real way the model will be used.

```python
#chose a lag
lag = 10

#explanatory variables
inputs_tv = dftv[:-24*lag]
# target variables to predict 
target_keys = ['mean_chla_depth']
targets_tv = dftv[24*lag:][target_keys]
#test input and targets:
inputs_test = dftest[:-24*lag]
targets_test = dftest[24*lag:]
#chose a model
model = linear_model(target_keys)


inputs_shuf = inputs_tv.sample(frac=1)
targets_shuf = targets_tv.loc[inputs_shuf.index+pd.Timedelta(lag,'D')]
```

```python
#run the kfold
n_fold = 4
#run the kfold (while keeping the same parameters as before)
res = kfold(model,inputs_shuf,targets_shuf,n_fold)
print("Model trained"+' '*70)
```

```python
# do the same plot as previously:


#lets only focus on a specific fold
fold_id  = 3
res_fold = res[fold_id]

#and lets only focus on a specific key
key_id  = 0

#plot the results##############
fig,ax = plt.subplots(1,figsize=(15,4))
true_data = targets_tv.iloc[:,key_id]
#plot the true data
true = ax.plot(true_data,c='k')

#plot the predicted training data
bp = res_fold['bp']
train = ax.scatter(targets_shuf.index[0:bp[fold_id]],res_fold['y_predtr'][0:bp[fold_id],key_id],c= 'tab:blue',s=1)
ax.scatter(targets_shuf.index[bp[fold_id+1]:],res_fold['y_predtr'][bp[fold_id]:,key_id],c='tab:blue',s=1)
#plot the validatation data
val = ax.scatter(targets_shuf.index[bp[fold_id]:bp[fold_id+1]],res_fold['y_predv'][:,key_id],c='tab:orange',s=1)

ax.legend([true[0],train,val],
         ['True data',
          f'Training predictions (R2={res_fold["s_train"]["R2_cat"][key_id]:.2f})',
          f'Validation predictions (R2={res_fold["s_val"]["R2_cat"][key_id]:.2f})'
         ])
ax.set_ylabel("Concentration [ROI/s]")
ax.set_title(f"fold number {fold_id} results for {target_keys[key_id]}")
plt.show()
```

A validation of 0.84 is quite a good result, but as soon as this model would be applied on a test dataset, it would drop heavily:

```python
# do the same plot as previously, but add the test prediction:
#lets only focus on a specific fold
fold_id  = 3
res_fold = res[fold_id]

#and lets only focus on a specific key
key_id  = 0

#plot the results##############
fig,ax = plt.subplots(1,figsize=(15,4))
true_data = targets_tv.iloc[:,key_id]
#plot the true data
true = ax.plot(true_data,c='k')

#plot the predicted training data
bp = res_fold['bp']
train = ax.scatter(targets_shuf.index[0:bp[fold_id]],res_fold['y_predtr'][0:bp[fold_id],key_id],c= 'tab:blue',s=1)
ax.scatter(targets_shuf.index[bp[fold_id+1]:],res_fold['y_predtr'][bp[fold_id]:,key_id],c='tab:blue',s=1)
#plot the validatation data
val = ax.scatter(targets_shuf.index[bp[fold_id]:bp[fold_id+1]],res_fold['y_predv'][:,key_id],c='tab:orange',s=1)
#plot the test data
pred_test = res_fold['model'].pred(inputs_test)
test = ax.scatter(targets_test.index,pred_test[:,key_id],s=1,c='tab:green')
true_test = ax.plot(targets_test.iloc[:,key_id],c='k')

ax.legend([true[0],train,val,test],
         ['True data',
          f'Training predictions (R2={res_fold["s_train"]["R2_cat"][key_id]:.2f})',
          f'Validation predictions (R2={res_fold["s_val"]["R2_cat"][key_id]:.2f})',
          f'Test predictions (R2={score(targets_test.to_numpy(),pred_test,res_fold["mean_ytrain"])["R2_cat"][key_id]:.2f})'
         ])
ax.set_ylabel("Concentration [ROI/s]")
ax.set_title(f"fold number {fold_id} results for {target_keys[key_id]}")
plt.show()
```

Thus an optimal batch size can be created, in such a way that the validation score is representative of the test one

```python
n_folds = [2,3,4,5,6,7,8,9,10,20,50,100,1000]

#chose a lag
lag = 10
#focus on a specific key:
key_id = 0
#explanatory variables
inputs_tv = dftv[:-24*lag]
# target variables to predict 
target_keys = ['mean_chla_depth']
targets_tv = dftv[24*lag:][target_keys]
#test input and targets:
inputs_test = dftest[:-24*lag]
targets_test = dftest[24*lag:]
#chose a model



val_res_n_folds = []
test_res_n_folds = []
res_n = []
for k in n_folds:
    model = linear_model(target_keys)
    res = kfold(model,inputs_tv,targets_tv,k)
    res_n.append(res)
    pred_test = [res_fold['model'].pred(inputs_test) for res_fold in res]
    score_test = [score(targets_test.to_numpy(),pred_test_i,res_fold["mean_ytrain"])["R2_cat"][key_id]
                  for pred_test_i,res_fold in zip(pred_test,res)]
    test_res_n_folds.append(score_test)
    val_res_n_folds.append([res_fold["s_val"]["R2_cat"][key_id] for res_fold in res])
```

```python
fig,ax = plt.subplots(1,figsize=(8,4))
mean_test = []
std_test = []
mean_val = []
std_val = []
for i in range(len(n_folds)):
    mean_test.append(np.mean(test_res_n_folds[i]))
    std_test.append(2*np.std(test_res_n_folds[i]))
    mean_val.append(np.mean(val_res_n_folds[i]))
    std_val.append(2*np.std(val_res_n_folds[i]))
x = np.arange(len(n_folds))
test = ax.bar(x,mean_test,width = 0.4,align='edge',yerr = std_test)
val = ax.bar(x,mean_val,width = -0.4,align='edge',yerr = std_val)
ax.legend([val,test],["validation results","test results"])
ax.set_xticks(x)
ax.set_xticklabels(n_folds)
ax.set_xlabel("Number of folds")
ax.set_ylabel("R2 score")
ax.set_ylim(-15,1)
plt.show()
```

```python
#plot the evolution of the test results vs number of folds:
fig,ax = plt.subplots(1,figsize=(6,4))
ax.plot(n_folds,mean_test)
ax.fill_between(n_folds, mean_test - 1.96*np.array(std_test), mean_test + 1.96*np.array(std_test),
                 color='tab:blue', alpha=0.2)
ax.plot(n_folds,mean_val)
ax.fill_between(n_folds, mean_val - 1.96*np.array(std_val), mean_test + 1.96*np.array(std_val),
                 color='tab:orange', alpha=0.2)
ax.set_ylim(-3,1)
ax.hlines(0,n_folds[0],n_folds[-1],color='k',linestyle='dashed')
plt.show()
```

As one can see on the graphs above, the test score is generally well represented by the validation score when the dataset is not shuffled, even for a relatively large number of folds. Moreover, one can note that the test dataset is (in the case of chlorophyll a) easier to fit than the mean fold of the training/validation set. Note that the reduction of the variance of the test score signifies that the models trained on the different folds are more and more alike as the number of fold increases, and not that the performance is improving. 

This analysis allows to use the number of folds to be chosen based only on the training time, without considering the apparently negligible higher correlation between the training set and the validation set, compared to the one of the training set and the test set

```python
#lets only focus on a specific fold of a specific test
fold_id  = 2
n_fold_id = 2
res_fold = res_n[n_fold_id][fold_id]

#and lets only focus on a specific key
key_id  = 0

#plot the results##############
fig,ax = plt.subplots(1,figsize=(15,4))
true_data = targets_tv.iloc[:,key_id]
#plot the true data
true = ax.plot(true_data,c='k')

#plot the predicted training data
bp = res_fold['bp']
train = ax.plot(true_data.index[0:bp[fold_id]],res_fold['y_predtr'][0:bp[fold_id],key_id],c= 'tab:blue')
ax.plot(true_data.index[bp[fold_id+1]:],res_fold['y_predtr'][bp[fold_id]:,key_id],c='tab:blue')
#plot the validatation data
val = ax.plot(true_data.index[bp[fold_id]:bp[fold_id+1]],res_fold['y_predv'][:,key_id],c='tab:orange')

pred_test = res_fold['model'].pred(inputs_test)
test = ax.plot(targets_test.index,pred_test[:,key_id],c='tab:green')
true_test = ax.plot(targets_test.iloc[:,key_id],c='k')


ax.legend([true[0],train[0],val[0],test[0]],
         ['True data',
          f'Training predictions (R2={res_fold["s_train"]["R2_cat"][key_id]:.2f})',
          f'Validation predictions (R2={res_fold["s_val"]["R2_cat"][key_id]:.2f})',
          f'Test predictions (R2={score(targets_test.to_numpy(),pred_test,res_fold["mean_ytrain"])["R2_cat"][key_id]:.2f})'
         ])
ax.set_ylabel("Concentration [ROI/s]")
ax.set_title(f"fold number {fold_id} results for {target_keys[key_id]}")
plt.show()
```
