# Datasets
The dataset, originally created to have a hold out approach, was modified in order to allow k-fold validation. This notebook is here to illustrate how it works

```python
import importlib
import os
import sys
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

stf_dir = os.path.join(os.pardir, 'spacetimeformer')
sys.path.append(stf_dir)
from spacetimeformer import data
importlib.reload(data) #to make sure the last version of stf is used
```

The previous dataset creator was parametrized the following way:
- A time series was given in a csv
- It was then split in 3 different parts (training, validation and testing) in a class called CSVTimeSeries
- At the same time, the data is standardized by using the mean and std of the training set
- This class has several properties described below


```python
#the target cols could be set manually, but it is easier to simply load the csv file and read them
data_path='../Datasets/Forecasting_Aqua/data_h_rolling_interpolated_clusters.csv'
df = pd.read_csv(data_path)
#the first column will then be the chla level
target_cols = ['mean_chla_depth']
target_cols+=list(df.keys().drop(['Datetime','mean_chla_depth']))

timeseries = data.CSVTimeSeries(data_path=data_path,
        target_cols=target_cols,
        read_csv_kwargs={},
        val_split = 0.15,
        test_split = 0.15,
        date_format = "%d/%m/%Y %H:%M",
        nanval=0,
    )
```

```python
##get the training data:
#timeseries.train_data
##get the validation data:
#timeseries.val_data
##get the test data:
timeseries.test_data
```

```python
#if one want the data back to scale:
normalized = timeseries.test_data
scaled_back = timeseries.reverse_scaling_df(normalized)

#plot:
fig,axs = plt.subplots(1,2,figsize=(15,4))
axs[0].plot(scaled_back['Datetime'],scaled_back['precipitation_tot'])
axs[1].plot(normalized['Datetime'],normalized['precipitation_tot'])
axs[0].set_title('Original data (restored)')
axs[1].set_title('Normalized data (stored)')
plt.show()
```

In addition to this scaler, the CSVTimeSeries object have a few other methods:

```python
#get a slice from start to end using a step of size k:
original = timeseries.val_data
start = 0
stop = -1 #meaning the last item
k = 100
slice_data = timeseries.get_slice('val',start,stop,k)


#plotting:
fig,ax = plt.subplots(1,figsize = (8,4))
ax.plot(original['Datetime'],original['mean_chla_depth'])
ax.scatter(slice_data['Datetime'],slice_data['mean_chla_depth'],c='orange')
ax.legend(['original data',f"sliced data every {k} hours"])
```

```python
# get the length of a split:
timeseries.length('val')
```

To allow to have a k-fold validation, a new class, CSVTimeSeriesV2 was created: It kept the all the utilities of the previous class, but instead of a validation split, one now have to specify how many times the data should be split (folds), and which of the splits should be used for validation (fold_n).

```python
folds = 10
fold_n = 5
timeseries2 = data.CSVTimeSeriesV2(data_path = data_path,
                             target_cols = target_cols,
                             folds = folds,
                             fold_n= fold_n,
                             test_split = 0.1,
                             date_format = "%d/%m/%Y %H:%M",
                             nanval=0,
                            )
```

This small modification allows to split the data more freely:

```python
#plot:
fig,axs = plt.subplots(1,2,figsize=(15,4))
for idx,series in enumerate([timeseries,timeseries2]):
    for split in ['train','val','test']:
        axs[idx].scatter(series.get_slice(split,0,-1,1)['Datetime'],series.get_slice(split,0,-1,1)['mean_chla_depth'],s=1)
    axs[idx].legend(['train','val','test'])
axs[0].set_title('Original class')
axs[1].set_title(f'New class with {folds} folds and using the {fold_n}th for validation')
plt.show()
```

The dataset creator had then to be slightly modified to allow this change, mainly the fact that the training data is now separated into 2 disjointed parts. It was also modified to allow two different time resolutions on the context and target part of each section:

```python
#create a dataset where the 128 points are taken 12 hours appart to predict 32 points, 10 days appart from one another

dset = data.CSVTorchDset(csv_time_series=timeseries2,
                        split= "train",
                        context_points = 128,
                        target_points = 32,
                        time_resolution_ctx = 12,
                        time_resolution_trg = 24*10
    )
```

```python
timeseries2.train_data['Datetime']
```

```python
idx = 0
x_ctx,y_ctx,x_trg,y_trg = dset[idx]
def x_to_date(x):
    x_scale = np.zeros_like(x,dtype=int)
    x_scale[:,0]=(((x[:,0]+1)/2)*30+1+0.1).int()
    x_scale[:,1]=(((x[:,1]+1)/2)*23+0.1).int()
    x_scale[:,2]=((x[:,2]+1)/2*11+1+0.1).int()
    x_scale[:,3]=((x[:,3]*2+2019)).int()
    #add all columns to get the relative time (in hours)
    x_date = np.array([np.datetime64(f"{row[3]}-{row[2]:02}-{row[0]:02}T{row[1]:02}") for row in x_scale])
    return x_date
x_ctx = x_to_date(x_ctx)
x_trg = x_to_date(x_trg)
```

```python
#plotting:
original = timeseries2.train_data
fig,ax = plt.subplots(1,figsize = (8,4))
ax.scatter(original['Datetime'],original['mean_chla_depth'],s=1)
ax.scatter(x_ctx,y_ctx[:,0],s=1,c='orange')
ax.scatter(x_trg,y_trg[:,0],s=1.5,c='red')
ax.legend(['original data',"context points",'target points'])
plt.show()
```
